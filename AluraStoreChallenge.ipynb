```python
# Função para formatar os ticks dos graficos
def formata_milhoes(x, pos):
    if x >= 1_000_000:
        return f"{x/1_000_000:.1f} mi"
    elif x >= 1_000:
        return f"{x/1000:.0f} mil"
    else:
        return int(x)
```

```python
#cores das barras
colors = [(0, 0, 1, 0.7),(0, 0, 1, 0.5), (0, 0, 1, 0.3), (0, 0, 1, 0.25)]


plt.figure(figsize=(8, 6))
ax = df_dados_faturamento.plot(kind='barh', x='Loja', y='Faturamento', color=colors)


# Aplicar formatação
ax.xaxis.set_major_formatter(mtick.FuncFormatter(formata_milhoes))
ax.set_ylabel('')

plt.xlabel("Faturamento", fontsize=10, fontweight= 'bold' )
plt.xlim(left=1_000_000)
plt.title("Faturamento total por Loja em R$", fontsize=14, fontweight= 'bold')
ax.axvline(media_faturamento, color='red', linestyle='-', linewidth=1.5, label=(f'Média R$ {media_faturamento:,.0f}'))
plt.grid(axis='x')
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

```

### ⭐ **Identificando a Categoria com Mais Faturamento**

Essa métrica determina o peso estratégico de cada categoria baseada no faturamento total da categoria considerando todas as lojas, portanto, a informação a ser coletada aqui é: 

***"Qual categoria é mais relevante para o negócio visando o faturamento".***


```python
df_total = pd.concat([loja1, loja2, loja3, loja4])

faturamento_categoria = (
    df_total.groupby('Categoria do Produto')['Preço']
    .sum()
    .sort_values(ascending=False)
)

faturamento_categoria
```

### **📊 Visualização Gráfica**

```python
# Preparar dados
categorias = faturamento_categoria.index
valores = faturamento_categoria.values

# Paleta moderna
cores = sns.color_palette("coolwarm", len(categorias))

# Criar figura e ax
fig, ax = plt.subplots(figsize=(10,6))

# Gráfico horizontal
bars = ax.barh(categorias, valores, color=cores)
ax.invert_yaxis()  # Maior faturamento no topo

# Adicionar valores no topo das barras
for bar in bars:
    width = bar.get_width()
    ax.text(width + 10_000, bar.get_y() + bar.get_height()/2,
            f'R$ {width:,.0f}', ha='left', va='center', fontsize=10, fontweight='bold')

# Ajustes via ax
ax.set_xlabel("Faturamento Total (R$)", fontsize=12, fontweight='bold')
ax.set_ylabel("Categoria do Produto", fontsize=12, fontweight='bold')
ax.set_title("Faturamento Total por Categoria (Todas as Lojas)", fontsize=14, fontweight='bold')
ax.grid(axis='x', linestyle='--', alpha=0.7)
ax.set_ylabel('')
ax.set_xlabel('')
ax.set_xticks([])

# alterando objeto ax
ax.spines['right'].set_visible(False)  # Remove a borda direita
ax.spines['top'].set_visible(False)    # Opcional: remove a borda superior também

# Exibir gráfico
plt.tight_layout()
plt.show()
```

### ⭐ **Identificando o Faturamento Total das Top 3 Categorias Lucrativas por Loja**

- Agora, identificar o faturamento Total dessas categorias por loja, determina o **lucro maior** delas, sendo fator determinante na somatória de valores, além de apoiar a **identificação** das lojas que mais **lucram** com as vendas das categorias **Top 3**.

```python
# Lista das top 3 categorias
top3_categorias = ['eletronicos', 'eletrodomesticos', 'moveis']

# Função para somar faturamento das top 3 categorias por loja
def faturamento_top3(data, loja_name):
    faturamento = data[data['Categoria do Produto'].isin(top3_categorias)]['Preço'].sum()
    return pd.DataFrame({'Loja': [loja_name], 'Faturamento Top 3': [faturamento]})

# Aplicar para cada loja
faturamento_loja1 = faturamento_top3(loja1, 'Alura Store 1')
faturamento_loja2 = faturamento_top3(loja2, 'Alura Store 2')
faturamento_loja3 = faturamento_top3(loja3, 'Alura Store 3')
faturamento_loja4 = faturamento_top3(loja4, 'Alura Store 4')

# Concatenar resultados
soma_top3_por_loja = pd.concat([faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4], ignore_index=True)

# Calcular a média
media_top3 = soma_top3_por_loja['Faturamento Top 3'].mean()

print("Faturamento das Top 3 categorias por loja:")
print(soma_top3_por_loja)
print(f"\nMédia do faturamento das Top 3 categorias por loja: R$ {media_top3:,.2f}")
```

### **📊 Visualização Gráfica**

```python
# Paleta moderna
cores = sns.color_palette("coolwarm", len(soma_top3_por_loja))

# Criar figura e eixo
fig, ax = plt.subplots(figsize=(10,6))
bars = ax.barh(soma_top3_por_loja['Loja'], soma_top3_por_loja['Faturamento Top 3'], color=cores)

# Adicionar valores no topo das barras
for i, v in enumerate(soma_top3_por_loja['Faturamento Top 3']):
    ax.text(v + 10000, i, f'R$ {v:,.0f}', va='center', fontsize=10, fontweight='bold')

# Adicionar linha da média (com legenda)
ax.axvline(media_top3, color='red', linestyle='--', linewidth=2, label=f'Média R$ {media_top3:,.0f}')

# Remover bordas desnecessárias
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Ajustes visuais
ax.set_title("Faturamento das Top 3 Categorias por Loja", fontsize=16, fontweight='bold')
ax.set_xlabel("Faturamento (R$)", fontsize=12, fontweight='bold')
ax.set_ylabel("")
ax.grid(axis='x', linestyle='--', alpha=0.7)

# Legenda apenas para a linha da média
ax.legend(loc='upper left',fontsize=10, title_fontsize=12)
ax.set_xticks([])

plt.tight_layout()
plt.show()
```

## 🔥 **Mapa de Calor de Vendas**

O **mapa de calor de vendas** vai permitir visualizar a intensidade das vendas em regiões e estados diferentes.

Com esse tipo de análise, é possível identificar **oportunidades de crescimento em estados, direcionamento de público, cultura a ser abordada e estratégias direcionadas a comportamento de clientes**.

```python
agrupado = (
    pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)
    .groupby(['lat', 'lon'], as_index=False)
    .agg(
        valor_total=('Preço', 'sum'),
        frequencia=('Preço', 'count')
    )
)

agrupado.sort_values(by='valor_total', ascending=False)
```

```python
import branca.colormap as cm

colormap = cm.linear.YlOrRd_09.scale(
    agrupado['valor_total'].min(),
    agrupado['valor_total'].max()
)
```

```python
import folium

mapa = folium.Map(location=[-14.2, -51.9], zoom_start=5)

for _, row in agrupado.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=row['valor_total'] ** 0.5 / 10,  # escala visual
        popup=f"""
        <b>Valor total:</b> R$ {row['valor_total']:,.2f}<br>
        <b>Vendas:</b> {row['frequencia']:.0f}
        """,
        fill=True,
        fill_color=colormap(row['valor_total']),
        fill_opacity=0.7,
        color=None
    ).add_to(mapa)

# adicionar legenda
colormap.caption = "Faturamento total por local"
colormap.add_to(mapa)

mapa
```
